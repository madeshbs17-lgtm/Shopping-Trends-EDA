```python
plt.hist(df['Age'])
plt.title("Histogram of Age")
plt.show()
plt.tight_layout()
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()
plt.tight_layout()
```

```python
df.groupby('Category')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
df['Subscription Status'].value_counts().plot(kind='pie')
plt.title("Subscription Status Distribution")
plt.show()
plt.tight_layout()
```

```python
df.groupby('Season')['Purchase Amount (USD)'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Total Purchase Amount by Season")
plt.show()
plt.tight_layout()
```

```python
df['Preferred Payment Method'].value_counts().sort_values(ascending=False)
```

```python
df.groupby('Size')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```

```python
df.groupby('Payment Method')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```

```python
df.groupby('Frequency of Purchases')['Previous Purchases'].sum().sort_values(ascending=False)
```

```python
df['Season'].value_counts().plot(kind='bar')
plt.title("Season Distribution")
plt.show()
plt.tight_layout()
```

```python
df['Category'].value_counts().plot(kind='pie')
plt.title("Category Distribution")
plt.show()
plt.tight_layout()
```

```python
df.boxplot(column='Purchase Amount (USD)',by = 'Frequency of Purchases')
plt.title("Purchase Amount by Frequency of Purchases")
plt.show()
plt.tight_layout()

```
